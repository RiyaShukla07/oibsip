```python
# Month of Unemployment
# These code converts the 'Date' column to a datetime type,extracts the month as an integer, and adds a new column
# With the corresponding three-letter month abbreviation.

import datetime as dt
import calendar

data['Date'] = pd.to_datetime(data['Date'],  dayfirst=True)
data['month_int'] = data['Date'].dt.month
data['month'] = (data['month_int'].astype(int)).apply(lambda x: calendar.month_abbr[x])
```

```python
data_11_2020['Date'] = pd.to_datetime(data_11_2020['Date'],  dayfirst=True)
data_11_2020['month_int'] = data_11_2020['Date'].dt.month
data_11_2020['month'] = data_11_2020['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
# visualize labor participation rate and unemployment rate in each month

data_EEE = data.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean()
data_EEE = pd.DataFrame(data_EEE).reset_index()
month = data_EEE.month
unemployment_rate = data_EEE['Estimated Unemployment Rate']
labor_participation_rate = data_EEE['Estimated Labour Participation Rate']

fig = go.Figure()

fig.add_trace(go.Bar(x = month, y = unemployment_rate, name = 'unemployment_Rate'))
fig.add_trace(go.Bar(x = month, y = labor_participation_rate, name = 'Labor Participation Rate'))

fig.update_layout(title='Unemployment Rate and Labor Participation Rate', xaxis={'categoryorder':'array', 'categoryarray':['Jan'],})
fig.show()
```

```python
data_EEE_11_2020 = data_11_2020.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean()
data_EEE_11_2020 = pd.DataFrame(data_EEE_11_2020).reset_index()
month = data_EEE_11_2020.month
unemployment_rate = data_EEE_11_2020['Estimated Unemployment Rate']
labor_participation_rate = data_EEE_11_2020['Estimated Labour Participation Rate']

fig = go.Figure()

fig.add_trace(go.Bar(x = month, y = unemployment_rate, name = 'unemployment_Rate'))
fig.add_trace(go.Bar(x = month, y = labor_participation_rate, name = 'Labor Participation Rate'))

fig.update_layout(title='Unemployment Rate and Labor Participation Rate', xaxis={'categoryorder':'array', 'categoryarray':['Jan'],})
fig.show()
```
